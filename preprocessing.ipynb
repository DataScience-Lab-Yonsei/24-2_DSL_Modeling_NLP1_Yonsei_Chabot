```python
import glob
import os
import re
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores.utils import DistanceStrategy
```

```python
file_pattern = "/text_706/following_texts_page_*.txt"
total_texts = []
file_list = glob.glob(file_pattern)

for i in range(len(file_list)):
    file_name = f"/text_706/following_texts_page_{i + 1}.txt"
    with open(file_name, "r", encoding="utf-8") as file:
        content = file.read()
        total_texts.append(content)
```

```python
len(total_texts)
```

```python
total_texts[3][:100]
```

```python
pattern_academic = r"학 \n"
pattern_engagement = r"Engagement \n"
pattern_innovative_leader = r"리더 \n"


processed_texts1 = []
#processed_texts1.append(total_texts[0])

for i in range(1, len(total_texts)):
    if i % 2 == 0:  # even index
        match = re.search(pattern_academic, total_texts[i])
        if not match:
            match = re.search(pattern_engagement, total_texts[i])
    else:  # odd index
        match = re.search(pattern_innovative_leader, total_texts[i])

    ind_newline = total_texts[i].rfind('\n')

    following_text = total_texts[i][match.end():ind_newline+2]
    processed_texts1.append(following_text)
```

```python
len(processed_texts1)
```

```python
processed_texts1[2]
```

```python
index_list = [10, 96, 109, 127, 164, 235, 256, 277, 285, 330, 345, 371, 399, 441, 625, 653, 672, 688, 698]

combined_texts = []

for i in range(len(index_list)):
    if i == 0:
        combined_text = ''.join(processed_texts1[:index_list[i]-1])
    elif index_list[i] == 441:
        continue
    else:
        combined_text = ''.join(processed_texts1[index_list[i-1]-1:index_list[i]-1])

    combined_texts.append(combined_text)


combined_text = ''.join(processed_texts1[index_list[-1]-1:])
combined_texts.append(combined_text)
```

```python
major_texts = []
for i in range(len(combined_texts)):
    majors = combined_texts[i].split('【')
    major_texts.extend(majors)
```

```python
len(major_texts)
```

```python
for i in range(len(major_texts)):
  print(major_texts[i][:60])
```

```python
major_texts1 = []
for text in major_texts:
    cleaned_text = text.replace("�", "")
    cleaned_text = text.replace("】", "")
    major_texts1.append(cleaned_text)
```

```python
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=2000,
    chunk_overlap=200,
    encoding_name='cl100k_base'
)
```

```python
total_text_split = []
for i in range(len(major_texts1)):
    current_split = text_splitter.split_text(major_texts1[i])
    total_text_split += current_split
```

```python
documents = []
i = 1
for custom_document in total_text_split:
    document = Document(page_content=custom_document, metadata={"source": i})
    documents.append(document)
    i+=1
```

3. Embedding

```python
embeddings_model = HuggingFaceEmbeddings(
    model_name='jhgan/ko-sroberta-nli',
    model_kwargs={'device':'cpu'},
    encode_kwargs={'normalize_embeddings':True},
)
```

4. Vector Store

```python
vectorstore = FAISS.from_documents(documents,
                                   embedding = embeddings_model,
                                   distance_strategy = DistanceStrategy.COSINE
                                  )
```

```python
vectorstore.save_local("/db/faiss")
```

Visualization of Vector Store

```python
embeddings_vector = vectorstore.index.reconstruct_n()
```

```python
embeddings_vector.shape
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Using t-SNE to reduce dimensions for visualization (reducing to 2D)
tsne = TSNE(n_components=2, random_state=42)
tsne_vectors = tsne.fit_transform(embeddings_vector)

# Plotting the t-SNE reduced vectors
plt.figure(figsize=(10, 7))
plt.scatter(tsne_vectors[:, 0], tsne_vectors[:, 1])
plt.title("t-SNE Visualization of VectorStore")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
